```python
# Task 1: write a code for Numeric and Boolean 
pos_int = 50
neg_int = -25
bin_int = 0b1101
oct_int = 0o17 
hex_int = 0x1A             

float_num = 3.14159

complex_num = 2 + 3j

is_true = True
is_false = False

print("Positive:", pos_int)
print("Negative:", neg_int)
print("Binary:", bin_int)
print("Octal:", oct_int)
print("Hexadecimal:", hex_int)

print("Float number:", float_num)

print("Complex number:", complex_num)
print("Real part:", complex_num.real)
print("Imaginary part:", complex_num.imag)

print("True value:", is_true)
print("False value:", is_false)
```

```python
# Task 2 : code for all arithmetic operators
a=15
b=4
print("Addition: a+b=", a+b)
print("Substraction: a-b=", a-b)
print("Multiplication: a*b=", a*b)
print("Division : a/b=", a/b)
print("floor division: a//b=", a//b)
print("modulus: a%b=", a%b)
print("exponential: a**b=", a**b)
```
