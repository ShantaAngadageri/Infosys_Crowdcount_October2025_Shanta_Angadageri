```python
# Task 1 
print("========= Google Form ========")
name=input("Enter your name:")
age=int(input("Enter Your age:"))
height=float(input("Enter your height"))        
complex_num=complex(input("Enter complex number"))
  
# 1. Using Comma
print("Name:", name, "Age:", age, "Height:", height, "Complex:", complex_num)
# 2. Using Plus
print("Name: " + name + " Age: " + str(age) + " Height: " + str(height) +
      "  Complex: " + str(complex_num))
# 3. Using Percent (%)
print("Name: %s  Age: %d  Height: %.2f  Complex: %s " %
      (name, age, height, complex_num))
# 4. Using .format()
print("Name: {}  Age: {}  Height: {:.2f}  Complex: {} ".format(
    name, age, height, complex_num))
# 5. Using f-string
print(f"Name: {name}  Age: {age}  Height: {height:.2f}  Complex: {complex_num} ")
                
```

```python
# Task 2: Functions manipulating lists
def add_item(mylist, item):
    mylist.append(item)
    return mylist
def remove_item(mylist, item):
    if item in mylist:
        mylist.remove(item)
    return mylist
def update_item(mylist, index, new_value):
    if 0 <= index < len(mylist):
        mylist[index] = new_value
    return mylist
def search_item(mylist, item):
    return item in mylist
def display_list(mylist):
    return mylist

mylist = ["apple", "banana", "cherry"]
print("Original List:", mylist)
add_item(mylist, "orange")
print("After Adding:", mylist)
remove_item(mylist, "banana")
print("After Removing:", mylist)
update_item(mylist, 1, "grape")
print("After Updating:", mylist)
print("Is 'apple' in list?", search_item(mylist, "apple"))
print("Final List:", display_list(mylist))
```

```python
# Task 3: Built-in functions for Dictionary
student = {
    "name": "Shanta",
    "age": 21,
    "course": "Computer Science"
}
print("Dictionary:", student)
print("Keys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())
print("Get age:", student.get("age"))
student.update({"age": 23, "grade": "A"})
print("After Update:", student)

# Pop
student.pop("course")
print("After Pop:", student)

# Length
print("Length of dictionary:", len(student))

# Clear
student.clear()
print("After Clear:", student)
```

```python
# Task 4: Functions to modify Set
def add_to_set(myset, item):
    myset.add(item)
    return myset
def remove_from_set(myset, item):
    myset.discard(item) 
    return myset
def update_set(myset, new_items):
    myset.update(new_items)
    return myset
def check_item(myset, item):
    return item in myset

myset = {"apple", "banana"}
print("Original Set:", myset)
add_to_set(myset, "cherry")
print("After Adding:", myset)
remove_from_set(myset, "banana")
print("After Removing:", myset)
update_set(myset, ["grape", "mango"])
print("After Updating:", myset)
#print("Is 'apple' in set?", check_item(myset, "apple"))
```

```python
# Task 5: Set Operations
set1 = {"apple", "banana", "cherry"}
set2 = {"banana", "cherry", "grape"}
print("Set1:", set1)
print("Set2:", set2)

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference (set1 - set2):", set1 - set2)
print("Symmetric Difference:", set1 ^ set2)
print("Is set1 subset of set2?", set1.issubset(set2))
print("Is set1 superset of set2?", set1.issuperset(set2))
```
