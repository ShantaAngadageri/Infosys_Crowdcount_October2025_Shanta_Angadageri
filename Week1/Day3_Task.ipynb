```python
# Task 1: Calculator with all 7 operators
print("===== Simple Calculator =====")
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, %, //, **): ")

if op == "+":
    print("Result =", a + b)
elif op == "-":
    print("Result =", a - b)
elif op == "*":
    print("Result =", a * b)
elif op == "/":
    print("Result =", a / b)
elif op == "%":
    print("Result =", a % b)
elif op == "//":
    print("Result =", a // b)
elif op == "":
    print("Result =", a ** b)
else:
    print("Invalid Operator")
```

```python
# Task 2: for loop examples with tuple, dictionary, and set
print("\nLoop through Tuple:")
my_tuple = (10, 20, 30, 40)
for item in my_tuple:
    print(item)
print("\nLoop through Dictionary:")
my_dict = {"name": "Shanta", "age": 21, "course": "CS"}
for key, value in my_dict.items():
    print(key, ":", value)
print("\nLoop through Set:")
my_set = {"apple", "banana", "mango"}
for fruit in my_set:
    print(fruit)
```

```python
# Task 3: Calculator using match-case (structural pattern matching)
print("\n===== Calculator with match-case =====")
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, %, //, **): ")

match op:
    case "+":
        print("Result =", a + b)
    case "-":
        print("Result =", a - b)
    case "*":
        print("Result =", a * b)
    case "/":
        print("Result =", a / b)
    case "%":
        print("Result =", a % b)
    case "//":
        print("Result =", a // b)
    case "":
        print("Result =", a ** b)
    case _:
        print("Invalid Operator")
```
