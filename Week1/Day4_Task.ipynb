```python
# Task 1: Recursive Fibonacci and Factorial

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        seq = fibonacci(n - 1)
        seq.append(seq[-1] + seq[-2])
        return seq

num = int(input("Enter a number: "))
print("Factorial of", num, "=", factorial(num))
print(f"Fibonacci sequence up to {num} terms:", fibonacci(num))
```

```python
# Task 2: Modules Examples
import os
import math
import random

print("Current Working Directory:", os.getcwd())
print("List of files in current directory:", os.listdir())

print("\nSquare root of 16:", math.sqrt(16))
print("Factorial of 5:", math.factorial(5))
print("Value of pi:", math.pi)

print("\nRandom integer between 1 and 10:", random.randint(1, 10))
print("Random float between 0 and 1:", random.random())
print("Random choice from list:", random.choice(['apple', 'banana', 'cherry']))
```

```python
# Task 3: Comprehensions and Runtime Arguments
squares = [x**2 for x in range(1, 6)]
print("\nList Comprehension (Squares):", squares)
cubes = tuple(x**3 for x in range(1, 6))
print("Tuple Comprehension (Cubes):", cubes)
num_dict = {x: x**2 for x in range(1, 6)}
print("Dictionary Comprehension:", num_dict)
def sum_all(*args):
    return sum(args)

print("\nSum using *args:", sum_all(10, 20, 30, 40))

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("\nDetails using **kwargs:")
print_details(name="Shanta", age=21, course="CS")
```
