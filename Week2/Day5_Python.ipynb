```python
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def multiply(x,y):
    return x*y
def divide(x,y):
    return x/y
def calculator():
    print("select operation")
    print("1. addd")
    print("2. substraction")
    print("3. multiply")
    print("4. Divide")
    choice=input("enter tour choice")
    num1=int(input("Enter first number"))
    num2=int(input("Enter second number"))
    if choice=='1':
        return add(num1,num2)
    elif  choice=='2':
        return sub(num1,num2)
    elif choice=='4':
        return multiply(num1,num2)
    elif choice=='4':
        return divide(num1,num2)
    else:
        print("invalid input")
calculator()
```
