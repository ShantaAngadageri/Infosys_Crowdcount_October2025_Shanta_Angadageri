```python
for i in range(1, 6):
    print(f"\nPerson {i}")
    weight = float(input("Enter weight in kg: "))
    height = float(input("Enter height in meters: "))

    bmi = weight / (height * height)

    print(f"BMI of Person {i} is: {bmi:.2f}")
```

```python
import numpy as np
heights = np.array([1.70, 1.60, 1.80, 1.55, 1.68])
weights = np.array([65, 70, 55, 50, 40])
bmi = weights / (heights ** 2)
bmi_rounded = np.round(bmi, 2)
for i in range(len(bmi_rounded)):
    print(f" BMI = {bmi_rounded[i]} of a student {i+1} ")
```
